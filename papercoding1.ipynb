```python
codinglist=[]

for kk in range(0,12):
        codinglist.append([])
  
        
        
```

```python
for kk in range(4,12):
     n=kk
     limit =(2*n-3) 
     codage=[]
     for i in range(n):
         codage.append([])
         for j in range(n):
             codage[i].append(0)

     for i in range(n):
        for j in range(n):
             row=2**i
             col=2**(j+n)
             codage[i][j]=row+col
             if (i==j) and (i==0 or i==(n-1)):
                   diagl=0
             else :
                   diagl=2**((i+j-1)+2*n)
             codage[i][j]= codage[i][j]+diagl
        
             if ((i==0) and (j==(n-1))) or ((j==0) and (i==(n-1))):
                     diagr=0
             else:
                  if (i<j):
                       x=(i-j)+limit
                  else:
                      x=(i-j)
                  diagr= 2**(x+2*n+limit)
             codage[i][j]= codage[i][j]+diagr
        
     l=[]
     for i in range(n):
         for j in range(n):
             l.append(codage[i][j])
     codinglist[n]=l
```

```python
 codinglist
```

```python
import pickle

# define a list of places
placesList = codinglist

with open('coding.data', 'wb') as filehandle:
    # store the data as binary data stream
    pickle.dump(placesList, filehandle)
```
